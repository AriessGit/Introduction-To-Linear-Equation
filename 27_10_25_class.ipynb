```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, Eq, solve
# Step 1: Define the functions
def linear_function1(x):
    return -1 * x + 5
def linear_function2(x):
    return 2 * x - 1
# Step 2: Find the intersection using sympy
x_sym = symbols('x')
eq1 = -x_sym + 5
eq2 = 2*x_sym - 1
solution = solve(Eq(eq1, eq2), x_sym)
x_intersect = float(solution[0])
y_intersect = linear_function1(x_intersect)
print(x_intersect)
print(y_intersect)
```
